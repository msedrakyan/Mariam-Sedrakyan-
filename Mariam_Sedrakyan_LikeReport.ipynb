##### For my hypothesis, I needed to sort my table by accounts and timestamp; this will help me later to understand what I liked most

<p> To get my timestamp, I had to pull out the "timestamp" in the "string_list_data" from my Json file "likes_media_likes." I repeated the steps below for my "Last year" like table</p>

```python
liked_timestemp=pd.DataFrame(like_report['likes_media_likes'])
like_time=[x['string_list_data'][0]['timestamp'] for x in like_report['likes_media_likes']]
liked_timestemp['timestamp']=like_time
liked_timestemp
```

___To test my hypothese I sorted my "like report" table from the most liked account to least liked account. 
Below you can see the top 10 accoutns I liked the most.___

```python
liked_timestemp.groupby('title').count().sort_values('timestamp',ascending = False).head(10)
```

```python
liked_last_year_timestemp=pd.DataFrame(like_report_last_year['likes_media_likes'])
like_last_year_time=[x['string_list_data'][0]['timestamp'] for x in like_report_last_year['likes_media_likes']]
liked_last_year_timestemp['timestamp']=like_last_year_time
```
